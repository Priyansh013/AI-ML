```python
import random
import numpy as np
```

# CREATING DATASET

```python
# Creating Dataset:
np.random.seed(5)
n = 100  # number of examples
X = []
for i in range(n):
    x = 5*i + 17 + np.random.randint(100)
    X.append([i,x])
        
        
X = np.array(X)
```

```python
X.shape
```

# CREATING MODEL

```python
n = X.shape[0]
n
```

```python
# Calculating sigma_x , sigma_s_sqrd , sigma_xy, sigma_x, sigma_y

sigma_x = np.sum(X,axis =0)[0:-1]
sigma_x
```

```python
sigma_y = np.sum(X,axis=0)[-1]
sigma_y
```

```python
sigma_x_sqrd = np.sum(np.multiply(X,X),axis=0)[0:-1]
sigma_x_sqrd
```

```python
sigma_xy = np.sum(X[:,0:-1] * X[:,-1].reshape(n,1))
sigma_xy
```

```python
print(sigma_x , sigma_y , sigma_xy , sigma_x_sqrd)
```

```python
# Fitting Model into y = ax+b
a = (n * sigma_xy - sigma_x*sigma_y)/(n*sigma_x_sqrd - (sigma_x)**2)
b = sigma_y/n - np.dot(a,sigma_x)/n
```

```python
a,b
```

```python
# Creating model:
def predict(x):
    ans = np.dot(a,x)+b
    return ans
```

# PLOTTING

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(X[:,0] , X[:,1],label="ACTUAL")
plt.plot([predict(i) for i in range(100)], label="PREDICTED")
plt.legend()
plt.show()
```
