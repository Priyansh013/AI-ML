IMPORTING LIBRARIES

```python
import numpy as np
import pandas as pd
import pickle
```

LOADING DATASET

```python
#function to convert data
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
file_path = r"/home/priyansh/Downloads/cifar-10-batches-py/data_batch_1"
data = unpickle(file_path)
file_path2 = r"/home/priyansh/Downloads/cifar-10-batches-py/test_batch"
test = unpickle(file_path2)
```

```python
data.keys()
```

```python
df = pd.DataFrame(data[b'data'])
labels = pd.DataFrame(data[b'labels'])
df
```

IMPLEMENTING MODEL

```python
def calculate_W(data,labels):
    W = []
    labels = np.array(labels)
    for i in range(0,10):
        selected_data = data[labels==i]   
        mean =np.mean(selected_data,axis=0)
        W.append(mean)
    
    W = pd.DataFrame(W)     
    return W
```

```python
W = calculate_W(df,labels)
```

Calculate Score:

```python
def calculate_score(W,image):
    score = np.dot(W,image)
    total = np.sum(score)
    return score/total
    
```

```python
def predict(image,df,labels):
    score = calculate_score(W,image)
    score = np.array(score)
    return np.argmax(score)
```

CHECK MODEL

```python
test_df = pd.DataFrame(test[b'data'])
test_labels = pd.DataFrame(test[b'labels'])
```

```python
count=0
for ind in range(test_df.shape[0]):
    image = test_df.iloc[ind]
    prediction = predict(image,df,labels)
    actual = test_labels.iloc[ind][0]
    print(f"Actual: {actual} , Prediction: {prediction}")
```
